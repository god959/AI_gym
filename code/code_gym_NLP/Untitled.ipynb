```python
#替換不需要的句子，並正規化
# start = time.clock()
gym_list =[]
for i in content_list:
    i = re.sub('※ 發信站: 批踢踢實業坊',' ', i)
    i = re.sub('※ 文章網址:',' ', i)
    i = re.sub('※ 編輯:',' ', i)
    i = re.sub('(臺灣)',' ', i)
    i = re.sub('來自:',' ', i)
    #只保留中文
    rule = re.compile(r"[^\u4e00-\u9fa5]")
    i = rule.sub('', i)
    gym_list.append(i)
```
